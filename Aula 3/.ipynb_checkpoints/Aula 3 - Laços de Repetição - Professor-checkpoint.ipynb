8. Vamos fazer um programa para verificar quem é o assassino de um crime. Para descobrir o assassino, a polícia faz um pequeno questionário com 5 perguntas onde a resposta só pode ser sim ou não:

a. Mora perto da vítima?

b. Já trabalhou com a vítima? 

c. Telefonou para a vítima?

d. Esteve no local do crime?

e. Devia para a vítima?

Cada resposta sim dá um ponto para o suspeito. A polícia considera
que os suspeitos com 5 pontos são os assassinos, com 4 a 3 pontos são
cúmplices e 2 pontos são apenas suspeitos, necessitando outras
investigações. Valores abaixo de 1 são liberados. 




```python
print('INTERROGATORIO!')
print('Informe sim ou não para as perguntas abaixo:')
resp_a = input('Você mora perto da vitima? ').lower()
resp_b = input('Você já trabalhou com a vítima? ').lower()
resp_c = input('Você já telefonou para a vítima? ').lower()
resp_d = input('Você esteve no local do crime? ').lower()
resp_e = input('Você devia para a vítima? ').lower()

pontos = 0

if resp_a == 'sim':
    pontos = pontos + 1
if resp_b == 'sim':
    pontos = pontos + 1
if resp_c == 'sim':
    pontos = pontos + 1
if resp_d == 'sim':
    pontos = pontos + 1
if resp_e == 'sim':
    pontos = pontos + 1
    
if pontos == 5:
    print('Você está preso meliante (é o assassino)')
elif pontos >= 3:
    print('Você está preso meliante (é o cumplice)')
elif pontos >= 2:
    print('Não confio em você (é o suspeito)')
else:
    print('Está liberado cidadão')
```

 9.  Um produto vai sofrer aumento de acordo com a Tabela 1 abaixo. Faça
um programa que peça para o usuário digitar o valor do produto de
acordo com o preço antigo e escreva uma das mensagens da Tabela
2, de acordo com o preço reajustado: 

![tabela_1](https://s3-sa-east-1.amazonaws.com/lcpi/ea3f2097-6c1a-4560-a632-d96508e81d09.png)


![tabela 2](https://s3-sa-east-1.amazonaws.com/lcpi/472a59db-16eb-4f65-a99a-df81a7dbed2f.png)



Será que conseguimos escrever o problema acima de uma forma mais fácil?


```python
preco = float(input('Digite o preco antigo do produto: '))

if preco <= 50:
    preco_atual = preco * 1.05
elif preco <= 100:
    preco_atual = preco * 1.10
elif preco <= 150:
    preco_atual = preco * 1.13
else:
    preco_atual = preco * 1.15
    
print('O preco recalculado é', preco_atual)

if preco_atual <= 80:
    print('Está barato')
elif preco_atual <= 115:
    print('Está razoavel')
elif preco_atual <= 150:
    print('Está normal')
elif preco_atual <= 170:
    print('Está caro')
else:
    print('Está muito caro')
```

```python
# não quero que o codigo aceite um salario menor que o salario minimo

salario = 0.0 #Criei um valor inicial que garante que a condicao é true no começo
while (salario < 1200): 
    salario = float(input('Digite o seu salário: ')) ## Dou uma chance da condição mudar para false
print('Você ganha R$', salario)
```

## Looping infinito

```python
## O looping infinito acontece quando NAO mudamos o valor da condição no nosso while

## O importante é SEMPRE se atentar as variaveis de controle
```

```python
contador = 0 # Serve para eu saber quantas voltas eu vou dar no código (loop)
soma = 0
qts_vai_somar = int(input('Digite quantos numeros quer somar: '))

while (contador < qts_vai_somar):
    numero = int(input('Digite um numero: '))
    soma = soma + numero
    contador = contador + 1 ## É importante que a variavel de controle mude DENTRO do loop (cuidado com identação)
    
print('A soma é ', soma)
```

**PONTO DE ATENÇÃO** 
Variáveis mudam ativamente de valor durante a execução de um código. Não lembrar disso dá varios nós na cabeça (principalmente pra quem ta começando agora)

**Dica** Se o python está te apresentando um código com a estrutura abaixo, lembre-se da ordem de execução

a = a + 1

1. Avalio a expressão que está a direita do igual
2. Atribuo o valor para a variável a esquerda do igual

Exemplo disso aqui -> Contador

```python
contador = 0 # Serve para eu saber quantas voltas eu vou dar no código (loop)
soma = 0

while (contador < 5):
    numero = int(input('Digite um numero: '))
    soma = soma + numero
    contador = contador + 1 ## É importante que a variavel de controle mude DENTRO do loop (cuidado com identação)
    
print('A soma é ', soma)
```
